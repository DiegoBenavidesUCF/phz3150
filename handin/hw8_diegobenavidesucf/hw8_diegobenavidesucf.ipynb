```python
print('Problem 3a:\n')

x = np.linspace( 12 , 46 , 100 )
#print(x)

y = np.sin((2*(np.pi)*x)/16) * np.cos((2*(np.pi)*x)/8)
#print(y)

z = np.cos(((np.pi)*x)/9) * np.sin(((np.pi)*x)/6)
#print(z)


#starting plot:
fig , ax = plt.subplots(figsize = (6,6))

#plotting:
ax.plot( x , y , linestyle = '-' , linewidth = 2 , color = 'r' , label = 'Y(x)' )
ax.plot( x , z , linestyle = '--' , linewidth = 2 , color = 'b' , label = 'Z(x)')

#labels:
ax.set_xlabel('X' , fontsize = 20)
ax.set_ylabel('Y & Z' , fontsize = 20)

#ticks:
ax.set_xticks( np.arange( 20 , 41 , 10) )

ax.set_xticklabels( np.arange( 20 , 41 , 10 ) , fontsize = 18 )

ax.set_yticks( np.arange( -1 , 1.1 , 0.5))
ax.tick_params( axis ='both', which ='both', length = 4 , direction = 'out' , labelsize = 18)

ax.set_xlim(xmin = 20)
ax.set_xlim(xmax = 40)

#legend:
ax.legend(fontsize = 18 , loc = 'upper left')

plt.show()
```
