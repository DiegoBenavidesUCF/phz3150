```python
print('Problem 2:\n')

fig = plt.figure(figsize = (10, 10))

ax1 = fig.add_subplot(1,1,1)

ax2 = ax1.twinx()

fig.suptitle('Comparing Actual Distances From the Sun and the Measured Distances', fontsize = 24)

observed_d = [ 0.38707637398311073 , 0.7233390285130801 , 1.5236961311031048 , 5.201400341240987 , 9.547745291596849 , 19.1827509207773 , 30.057733077422224]
actual_d = [ 0.4 , 0.7 , 1.524 , 5.2 , 9.6  , 19.2 , 30.1 ]

period_in_years = [ 87.96 , 224.7 , 686.97 , 4332.82 , 10775.6 , 30687.15 , 60190.03 ]

#for actual distances:

ax1.plot(period_in_years , actual_d , linestyle = 'none' , label = 'Actual Distances' , color = 'green' , marker = 'o' , markersize = 16)
ax1.set_xlabel( ' Period (Earth Years) ' , fontsize = 16 )
ax1.set_ylabel( ' Actual Distance (AU) ' , color = 'green',fontsize = 16 )
#ax1.set_xticklabels( [  '0', 'mercury' , 'venus' , 'mars' , 'jupiter' , 'saturn' , 'uranus' , 'neptune' ] )

# for observed distances:

ax2.plot(period_in_years , observed_d , linestyle = 'none' , label = 'Observed Distances' , color = 'red' , marker = 'x' , markersize = 16) 
ax2.set_xlabel( ' Period (Earth Years)' , fontsize = 16 )
ax2.set_ylabel( ' Observed Distance (AU) ' , color = 'red', fontsize = 16 )
#ax2.set_xticklabels( [ '0' ,'mercury' , 'venus' , 'mars' , 'jupiter' , 'saturn' , 'uranus' , 'neptune' ] )

ax1.legend(loc = 'upper left' , fontsize = 12)
ax2.legend(loc = 'center left' , fontsize = 12)
```
